# Statistics Basic - Mean

```python
# the mean is the average of a set of number it is calculated by summing all the numbers and dividing by the numbers and dividing by the count of numbers

# use case the mean is used in marked analysis to summarize the average value of a dataset such as average customer age or average income
```

```python
import pandas as pd

# sample data
data={
    'Age':[25,30,35,40,45,50],
    'salary':[50000,60000,70000,80000,90000,30000]
}
df=pd.DataFrame(data)
# Mean
mean_age=df['Age'].mean()
mean_salary=df['salary'].mean()

print("Age Mean",mean_age)
print("Salary means",mean_salary)
```

```python
# Median
median_age=df['Age'].median()
median_salary=df['salary'].median()

print("Age Median",median_age)
print("Salary Median",median_salary)
```

```python
# Even median
data={
    'age':[2,4,6,8]
}
df=pd.DataFrame(data)
median=df['age'].median()
print(median)
```

```python
print(df)
```

```python
#odd median

data={
    'age':[1,2,3]
}
df=pd.DataFrame(data)
median=df['age'].median() # (n+1)/2
print(median)
```

# Descriptive Satistics Mode

```python
# Mode
data={
    'age':[2,4,6,8,2,4,6,8]
}
df=pd.DataFrame(data)
mode=df['age'].mode()
print(mode)
```

```python
# more than one mode

#multimodel data set

data={
    'age':[2,4,6,8,2,4,6,8,2,4,6,8]
}
df=pd.DataFrame(data)
mode=df['age'].mode()[0]
print(mode)
```

# Descriptive Statistics -Standard Deviation

```python
# Standard deviation

data={
    'age':[2,4,6,8,2,4,6,8]
}
df=pd.DataFrame(data)
std=df['age'].std()
print(std)
```

# Descriptive Statistics -Variance

```python
# variance

var_age=df['age'].var()
print(var_age)
```

# Probability Distributions Normal Distributions
# (Gaussian Distrbution)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the normal distribution
mu, sigma = 0, 0.1

# Generate points on the x axis
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)

# Plot the normal distribution
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

# Probability Distribution -Binomial Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the binomial distribution
n = 10   # number of trials
p = 0.5  # probability of success

# Generate possible number of successes
x = np.arange(0, n + 1)

# Calculate the binomial probability mass function (pmf)
pmf = stats.binom.pmf(x, n, p)

# Plot the binomial distribution
plt.bar(x, pmf, color='blue', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```
