# Correlataion

```python
import numpy as np
import pandas as pd

# seed for reproducibility

np.random.seed(42)

# generate sample data

data={
    'Age':np.random.normal(30,10,100).astype(int),
    'Weight':np.random.normal(50,10,100).astype(int),
    'Height':np.random.normal(60,10,100).astype(int)
}
# create Dataframe

df=pd.DataFrame(data)
df
```

```python
!pip install tabulate
```

```python
from tabulate import tabulate
```

```python
correlations=df.corr()
correlations
```

```python
# print the correlation matrix as a table
print(tabulate(correlations, headers='keys', tablefmt='grid',numalign='right',floatfmt='.2f'))
```
