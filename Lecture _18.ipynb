```python
import matplotlib.pyplot as plt
import numpy as np


#Data
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# creating line plot
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
#Data
x=[1,2,3,4,4,5,6,7,8,9]
y=[50,51,52,48,47,46,45,44,43,42]
plt.plot(x,y,marker='o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.show
```

```python
x=[1,2,3,4,4,5,6,7,8,9]
y=[50,51,52,48,47,46,45,44,43,42]

plt.plot(x,y,color='green',linestyle='dashed',linewidth=3,marker='*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.show
```

# Creating a Simple Scatter Plot with Matplotlib

```python
# Data
x = np.array([5,7,8,7,2,17,2,9,4,11])
y = np.array([99,86,87,88,111,86,103,87,94,78])
plt.scatter(x,y,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Scatter Plot')
plt.show()
```

# Creating a Simple Bar Plot with Matplotlib

```python
# Data
product=['a','b','c','d']
profit=[100,200,300,400]
colors=['red','green','#E07D0B','blue']
plt.bar(product,profit,color=colors,width=0.3)
plt.xlabel('product')
plt.ylabel('profit')
plt.title('Simple Bar Plot')
plt.show()
```

```python
# Data
product=['a','b','c','d']
profit=[100,200,300,400]
colors=['red','green','#E07D0B','blue']
plt.barh(product,profit,color=colors,height=0.3) # height for horizotal bar graph
```

# Creating a Simple histogram

```python
# Data
data=[1,2,3,4,5,6]
plt.hist(data,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Histogram')
plt.show()
```

# creating subplots with matplotlibs

```python
# data
x=[1,2,3,4,5]
y1=[2,3,5,8,9]
y2=[12,20,35,40,50]

# creating subplot
fig ,axs=plt.subplots(2,1)
axs[0].plot(x,y1)
axs[0].set_title('First Plot')
axs[1].plot(x,y2,'tab:orange')
axs[1].set_title('Second Plot')
# axs[1].plot(x,y2)
plt.tight_layout()
plt.show()
```

# Adding Annotations with matplotlib

```python
# Data
x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot with Annotation')
plt.annotate('peak',xy=(5,11),xytext=(4,10),arrowprops=dict(facecolor='red',shrink=0.05))
plt.show()
```

#Creating a Simple line plot with seaborn

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
x=[1,2,3,4,5]
y=[2,3,5,7,11]
sns.lineplot(x=x,y=y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Plot')
plt.show()
```

# Creating a simple scatter plot with seaborn

```python
#Data
x=[1,2,3,4,5]
y=[2,4,8,5,9]
sns.scatterplot(x=x,y=y)
sns.show()
```
