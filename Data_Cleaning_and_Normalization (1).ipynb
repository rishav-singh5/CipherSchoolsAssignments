```python
# prompt: create a data frame with 10 row and 4 columns that given columns nameName  Age  Salary   Department name be also random

import pandas as pd
import numpy as np

# Generate random data
names = ['Alice', 'Bob', 'Charlie', 'David', 'Emily', 'Frank', 'Grace', 'Henry', 'Ivy', 'Jack']
ages = np.random.randint(20, 60, 10)
salaries = np.random.randint(20000, 100000, 10)
departments = ['HR', 'Engineering', 'Sales', 'Marketing']
departments = np.random.choice(departments, 10)

# Create DataFrame
data = {
    'Name': names,
    'Age': ages,
    'Salary': salaries,
    'Department': departments
}
df_random = pd.DataFrame(data)

print(df_random)
```
